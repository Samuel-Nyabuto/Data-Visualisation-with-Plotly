```python
# Import pandas for data manipulation
import pandas as pd

# Import numpy for scientific computations
import numpy as np

# Import plotly library for data visualisation
# ---
# `plotly.express` contains plotly.py's core functionality
# ----
#
import plotly.express as px
```

Data Understanding:

```python
# Loading data:
mobile_df = pd.read_csv('https://bit.ly/MobileDataset')
mobile_df.head()
```

```python
mobile_df.shape
```

```python
mobile_df.dtypes
```

```python
# check for missing values in each column 
mobile_df.isnull().sum()
```

```python
# Inserting an underscore between column names for easy reading.
mobile_df.columns = mobile_df.columns.str.strip().str.replace(' ', '_').str.replace('[','').str.replace(']','')
mobile_df.head()
```

```python
# Checking for duplicates;
mobile_df[mobile_df.duplicated()]
sum(mobile_df.duplicated())
```

```python
# Dropping rows with missing values;
df = mobile_df.dropna()
df.sample(5)
```

```python
columns_df = list(mobile_df.columns[1:])
G_df = df.groupby('Series_Name', as_index=False).sum()
G_df.head()
```

```python
G_df = G_df.transpose().reset_index()
G_df
```

```python
new_df  = G_df.iloc[0]
new_df
```

```python
G_df.columns = new_df
G_df.drop(index=0, inplace=True)
G_df.head(2)
```

```python
new_columns = G_df.columns[1:]
G_df = pd.melt(G_df, id_vars=['Series_Name'], value_vars='value_columns')
G_df.head(5)
```

```python
fig = px.line(
    G_df,
    x = "Series_Name",
    y = "value",
   color = 0,
)
fig.update_layout(
    title = {
        'text': "Mobile Evolution Trends Over the Years",
        'x': 0.5,
        'font_size' : 15,
        'xanchor': "center",
        'yanchor': "auto"
    },
xaxis_title_text = "Years",
yaxis_title_text = "Users",
)

#Plot display

fig.show()
```

```python
yearly_columns = list(G_df.columns[2:])
G_df = pd.melt(G_df, id_vars=['Series_Name'], value_vars='country_name')
cellular_df = G_df[G_df['Series_Name']=='Fixed broadband subscriptions']

fig2 = px.line(
    cellular_df,
    x = "variable",
    y = "value",

  color = 'country_name'
)


# Tweak and Label
fig2.update_layout(
    title = {
        'text': "Fixed Broadband Per Country Over the Years",
        'x': 0.5,
        'font_size' : 15,
        'xanchor': "center",
        'yanchor': "auto"
    },
xaxis_title_text = "Years",
yaxis_title_text = "Users",
)

#Plot display

fig2.show()
```

```python
mobile_cell_df = G_df[G_df['Series_Name']=='Mobile cellular subscriptions']
fig3 = px.line(
    mobile_cell_df,
    x = "variable",
    y = "value",
    color = "country_name"
)

# Tweak and Label

fig3.update_layout(
    title = {
        'text': "Cellular Subscriptions per Country Over Years ",
        'x': 0.5,
        'font_size' : 15,
        'xanchor': "center",
        'yanchor': "top"
    },
    xaxis_title_text = "Years",
yaxis_title_text = "No of Cellular Subscrioptions",
)

# Display

fig3.show()
```
